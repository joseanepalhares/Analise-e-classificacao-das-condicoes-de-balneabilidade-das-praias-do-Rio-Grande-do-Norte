# 5) Vizualização dos dados


```
import matplotlib.pyplot as plt

labels = 'Nizia Floresta', 'Natal', 'Parnamirim', 'Extremoz'
sizes = [15, 30, 45, 10] #modificar os valores depois
explode = (0, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

ax1.axis('equal')
plt.title('Indice de contaminação média por cidade')
plt.show()
```

```
import matplotlib.pyplot as plt

idxes = [1,  2,  3,  4,  5,  6,  7,  8,  9]
arr1  = [540, 920, 3500, 5400,  1600, 3500, 540, 920, 49]
arr2  = [240, 350, 540, 3500, 1400, 2400, 130, 2100, 540]
arr3  = [70, 70, 7, 49, 240, 49, 110, 3500, 23]

plt.plot([], [], color='r', label = 'NF-05 (foz do rio pirangi) Nízia Floresta')
plt.plot([], [], color='g', label = 'NA-13 (redinha-rio potengi) Natal')
plt.plot([], [], color='b', label = 'EX-07 (lagoa de pitangi)Extremoz')

plt.stackplot(idxes, arr1, arr2, arr3, colors= ['r', 'g', 'b'])
plt.title('Indice de contaminação media por cidade')
plt.legend()
plt.show()
```
